# Task 2: Visualizing Mathematical Functions
1) Plot the following function:
$$
f(x) = e^{-x^2 / 3} + 3\sin\left(x + \frac{\pi}{4}\right)
$$

```python
%pip install numpy matplotlib
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Create an array of t values
t = np.linspace(0, 2 * np.pi, 500)

# Compute x(t) and y(t)
x_t = 2 * np.sin(t) * np.exp(np.cos(t))
y_t = -1.5 * np.cos(t) * np.exp(np.sin(2 * t))

# Plot the parametric curve
plt.figure(figsize=(8, 6))
plt.plot(x_t, y_t, label=r'Parametric curve: $x(t), y(t)$', color='darkred')
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.title("Parametric Plot of x(t) and y(t)")
plt.grid(True)
plt.legend()
plt.axis('equal')  # To maintain aspect ratio
plt.show()
```

2) Plot the following function 𝑓(𝑥):

$$
x(t) = 2\sin(t)\cdot e^{\cos(t)}
$$

$$
y(t) = -\frac{3}{2}\cos(t)\cdot e^{\sin(2t)}
$$

Where:

$$
0 \leq t \leq 2\pi
